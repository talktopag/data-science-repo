```python
import pandas as pd
import numpy as np
import plotly_express as px
from sqlalchemy import create_engine

pd.options.plotting.backend = 'plotly'

cnx = create_engine('sqlite:///assets/operations.db')
df = pd.read_sql('SELECT * FROM qc_test', cnx)
```

```python
lots = df.lot.unique()[-20:]

fig = df.plot(
    kind = 'strip', 
    x = 'lot', 
    y = 'result', 
    facet_row = 'level',
    height = 800,
    width = 800,
    labels = {'lot': 'Device Lot Number', 'result': 'Optical Density'},
    title = 'LFT QC Testing History'
)
fig.update_xaxes(type='category', tickangle = 90)
fig.update_yaxes(matches = None)
fig.show()
```
